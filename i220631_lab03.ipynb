# Lab Task 01

Reshape a 1D NumPy array into a 2D array with dimensions (3, 4).

Concatenate two NumPy arrays horizontally and vertically.

```python
import numpy as np
arr1=np.arange(12)
arr1=arr1.reshape(3,4)
print('\n\n',arr1)
arr2=np.array([[9,8,7,6],
               [3,4,5,6],
               [1,3,2,9]])
conc_h=np.concatenate((arr1,arr2),axis=1)
print('\n\n',conc_h)

conc_v=np.concatenate((arr1,arr2),axis=0)
print('\n\n',conc_v)
```

# Lab Task 02


## Email Classification
Develope a tool which will classify an email.

### Tasks:
* Create a function classify_emails that takes a list of emails as an argument.
* Inside the function, categorize emails into "Important," "Promotions," and "Spam" based on keywords and content analysis.
* Loop through the emails and identify keywords to determine their category.
* If an email contains keywords indicating importance, assign it as "Important."
* If an email contains keywords related to promotions or deals, assign it as "Promotions."
* If an email is suspected to be spam based on certain keywords or patterns, assign it as "Spam."
* Return three lists of emails for each category.
* Print the lists.

```python
#code here
def classify_emails(lists_email):
  important=[]
  promotions=[]
  spam=[]
  for email in lists_email:
    if 'important' in email:
      important.append(email)
    if 'promotion' in email or 'deal' in email:
      promotions.append(email)
    if 'spam' in email or 'fake' in  email:
      spam.append(email)
  print('Important Emails', important)
  print('Promotions Emails', promotions)
  print('Spam Emails', spam)
list_emails=[]
print("Enter q to quit")
data=input('\nEnter your email: ')
while data != 'q':
  list_emails.append(data)
  data=input('Enter your email: ')
print('\n')
classify_emails(list_emails)
```

# Lab Task 03

## English-Spanish Dictionary
You are creating a basic English-Spanish dictionary that allows users to look up English words and find their Spanish translations.


### Tasks
* Create a dictionary where English words are keys and their Spanish translations are values.
* Ask the user to input an English word.
* Check if the word exists in the dictionary.
* If the word exists, print its Spanish translation.
* If the word does not exist, inform the user that the translation is not available.
* Implement a loop that allows the user to keep looking up words until they choose to exit.


```python
#code here
spanish_dict = {
    "elections": "elecciones",
    "party": "fiesta",
    "holidays": "vacaciones",
    "hello": "hola",
    "enjoying": "disfrutando",
    "traveling": "viajando",
}

def translate_word(word):
    if word in spanish_dict:
        return spanish_dict[word]
    else:
        return "Translation not available"

while True:
    word = input("Enter an English word (or 'q' to quit): ").lower()

    if word == 'q':
        print("Goodbye!")
        break

    translation = translate_word(word)
    print("Translation: ", translation)
```

Use the dataset of Game, we have explored earlier and perform following operations.

* Calculate the total global sales for all games in the dataset.
* Find the game with the highest global sales.
* Determine the average sales for each region (NA_Sales, EU_Sales, JP_Sales, Other_Sales).
* Find the top 5 most popular game genres based on global sales.
* Calculate the total sales for each genre in each region and find the most popular genre in North America (NA_Sales).

* Determine the top 3 gaming platforms with the highest global sales.
* Calculate the market share (percentage of global sales) for each genre.
* Calculate the correlation coefficient between NA_Sales and EU_Sales. (.corr function)

```python
import pandas as pd
data=pd.read_csv('https://raw.githubusercontent.com/TirendazAcademy/PANDAS-TUTORIAL/main/DataSets/vgsalesGlobale.csv')
data
```

```python
total_global_sales = data['Global_Sales'].sum()
total_global_sales
```

```python
max_index=data['Global_Sales'].idxmax()
data['Name'][max_index]
```

```python
NA_avg=data['NA_Sales'].mean()
EU_Sales_avg=data['EU_Sales'].mean()
JP_Sales_avg=data['JP_Sales'].mean()
Other_Sales_avg=data['Other_Sales'].mean()
print("Average sales are ")
print("NA sales average: ",NA_avg)
print("EU sales average: ",EU_Sales_avg)
print("JP sales average: ",JP_Sales_avg)
print("Other sales average: ",Other_Sales_avg)
```

```python
genre_sales = data.groupby('Genre')['Global_Sales'].sum()
sorted_genre_sales = genre_sales.sort_values(ascending=False)
top_5_genres = sorted_genre_sales.head(5)
print("Top 5 most popular game genres based on global sales:")
print(top_5_genres)
```

```python
genre_sales_na = data.groupby('Genre')['NA_Sales'].sum()
genre_sales_eu = data.groupby('Genre')['EU_Sales'].sum()
genre_sales_jp = data.groupby('Genre')['JP_Sales'].sum()
genre_sales_other = data.groupby('Genre')['Other_Sales'].sum()

genre_region_sales = pd.DataFrame({
    'NA_Sales': genre_sales_na,
    'EU_Sales': genre_sales_eu,
    'JP_Sales': genre_sales_jp,
    'Other_Sales': genre_sales_other
})

print("Total sales for each genre in each region:")
print(genre_region_sales)
most_popular_genre_na = genre_region_sales['NA_Sales'].idxmax()

print("\nMost popular genre in North America:", most_popular_genre_na)
```

```python
data[len(data)-10:]
```

```python
genre_sales = data.groupby('Platform')['Global_Sales'].sum()
sorted_genre_sales = genre_sales.sort_values(ascending=False)
top_platforms = sorted_genre_sales.head(3)
print("Top 3 most popular platform based on global sales:")
print(top_platforms)
```

```python
genre_global_sales = data.groupby('Genre')['Global_Sales'].sum()
total_global_sales = data['Global_Sales'].sum()
market_share = (genre_global_sales / total_global_sales) * 100
print("Market share :")
print(market_share)
```

```python
corr_coeff = data['NA_Sales'].corr(data['EU_Sales'])
corr_coeff
```
